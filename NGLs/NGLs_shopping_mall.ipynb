```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the nodes with their names, tickers, and chemical compounds
nodes = {
    "Natural Gas\n(NG)\n(CH)": {"color": "yellow", "pos": (0, 8)},
    "Ethane\n(CAP)\n(CH)": {"color": "green", "pos": (-6, 4)},
    "Propane\n(BAP)\n(CH)": {"color": "green", "pos": (-3, 4)},
    "Butane\n(DAE)\n(CH)": {"color": "green", "pos": (0, 4)},
    "Isobutane\n(IBD)\n(CH)": {"color": "green", "pos": (3, 4)},
    "Natgas Complex\n(MMB)\n(CH)": {"color": "green", "pos": (6, 4)},
    "Ethylene\n(PCW)\n(CH)": {"color": "lightblue", "pos": (-4.5, 0)},
    "Propylene\n(PGP)\n(CH)": {"color": "lightblue", "pos": (1.5, 0)}
}

edges = [
    ("Natural Gas\n(NG)\n(CH)", "Ethane\n(CAP)\n(CH)"),
    ("Natural Gas\n(NG)\n(CH)", "Propane\n(BAP)\n(CH)"),
    ("Natural Gas\n(NG)\n(CH)", "Butane\n(DAE)\n(CH)"),
    ("Natural Gas\n(NG)\n(CH)", "Isobutane\n(IBD)\n(CH)"),
    ("Natural Gas\n(NG)\n(CH)", "Natgas Complex\n(MMB)\n(CH)"),
    ("Ethane\n(CAP)\n(CH)", "Ethylene\n(PCW)\n(CH)"),
    ("Propane\n(BAP)\n(CH)", "Ethylene\n(PCW)\n(CH)"),
    ("Propane\n(BAP)\n(CH)", "Propylene\n(PGP)\n(CH)"),
    ("Butane\n(DAE)\n(CH)", "Propylene\n(PGP)\n(CH)"),
    ("Isobutane\n(IBD)\n(CH)", "Propylene\n(PGP)\n(CH)")
]

# Create directed graph
G = nx.DiGraph()

# Add nodes with attributes
for node, attr in nodes.items():
    G.add_node(node, color=attr["color"], pos=attr["pos"])

# Add edges
G.add_edges_from(edges)

# Extract positions and colors
positions = {node: attr["pos"] for node, attr in nodes.items()}
node_colors = [attr["color"] for _, attr in G.nodes(data=True)]

# Set the figure size for clarity
plt.figure(figsize=(16, 12))

# Draw the nodes as larger ovals
nx.draw_networkx_nodes(G, positions, node_color=node_colors, node_size=12000, alpha=0.8, node_shape='o')

# Draw the edges with arrows
nx.draw_networkx_edges(G, positions, arrowstyle="->", arrowsize=20, edge_color='gray')

# Draw the labels with adjusted font size
nx.draw_networkx_labels(G, positions, font_size=12, font_color="black")

# Set the plot title
plt.title("Hierarchical Real Optionality Tree for NGLs and Petrochemicals")
plt.axis('off')
plt.show()
```
